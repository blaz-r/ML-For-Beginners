```python
import pandas as pd

df = pd.read_csv("../data/Hotel_reviews.csv")
```

Drop lat and lng

```python
df = df.drop(["lat", "lng"], axis=1)

df.info()
```

Replace Hotel_Address values with the following values (if the address contains the same of the city and the country, change it to just the city and the country).

These are the only cities and countries in the dataset:

Amsterdam, Netherlands

Barcelona, Spain

London, United Kingdom

Milan, Italy

Paris, France

Vienna, Austria

```python
def replace_address(row):
    if "Netherlands" in row["Hotel_Address"]:
        return "Amsterdam, Netherlands"
    elif "Barcelona" in row["Hotel_Address"]:
        return "Barcelona, Spain"
    elif "United Kingdom" in row["Hotel_Address"]:
        return "London, United Kingdom"
    elif "Milan" in row["Hotel_Address"]:        
        return "Milan, Italy"
    elif "France" in row["Hotel_Address"]:
        return "Paris, France"
    elif "Vienna" in row["Hotel_Address"]:
        return "Vienna, Austria" 

# replace with shorter addresses
df["Hotel_Address"] = df.apply(replace_address, axis=1)
```

```python
print(df["Hotel_Address"].value_counts())
```

```python
display(df.groupby("Hotel_Address").agg({"Hotel_Name": "nunique"}))
```

Drop Additional_Number_of_Scoring

```python
df = df.drop("Additional_Number_of_Scoring", axis=1)

df.info()
```

Replace Total_Number_of_Reviews with the total number of reviews for that hotel that are actually in the dataset

Replace Average_Score with our own calculated score

```python
# count all reviews for certain hotel
df["Total_Number_of_Reviews"] = df.groupby("Hotel_Name").Hotel_Name.transform("count")

# calcualte average score
df["Average_Score"] = df.groupby("Hotel_Name").Reviewer_Score.transform("mean")

df.head()
```

Process review columns

    Drop Review_Total_Negative_Word_Counts, Review_Total_Positive_Word_Counts, Review_Date and days_since_review

    Keep Reviewer_Score, Negative_Review, and Positive_Review as they are,

    Keep Tags for now

    We'll be doing some additional filtering operations on the tags in the next section and then tags will be dropped

```python
df = df.drop(["Review_Total_Negative_Word_Counts", "Review_Total_Positive_Word_Counts", "Review_Date", "days_since_review"], axis=1)
```

```python
df.info()
```

Process reviewer columns

Drop Total_Number_of_Reviews_Reviewer_Has_Given

Keep Reviewer_Nationality

```python
df = df.drop("Total_Number_of_Reviews_Reviewer_Has_Given", axis=1)
```

```python
df.info()
```

```python
df.Tags = df.Tags.str.strip("[']")

df.Tags = df.Tags.str.replace(" ', '", ",", regex = False)

df.Tags
```

```python
split_df = df.Tags.str.split(',', expand = True)
```

```python
# split tags into 6 columns
df["Tag_1"] = split_df[0].str.strip()
df["Tag_2"] = split_df[1].str.strip()
df["Tag_3"] = split_df[2].str.strip()
df["Tag_4"] = split_df[3].str.strip()
df["Tag_5"] = split_df[4].str.strip()
df["Tag_6"] = split_df[5].str.strip()
```

```python
# join six columns into one
df_tags = df.melt(value_vars=["Tag_1", "Tag_2", "Tag_3", "Tag_4", "Tag_5", "Tag_6"])
```

```python
df_tags.value_counts()
```

```python
# drop all which contain folloing words
df_tags = df_tags[~df_tags.value.str.contains("Standard|room|Stayed|device|Beds|Suite|Studio|King|Superior|Double", na=False, case=False)]
# keep all that have at least 1000 count
tag_vc = df_tags.value.value_counts().reset_index(name="count").query("count > 1000")
```

```python
tag_vc
```

```python
# create new columns in original df, fill with 1 or 0 accoridng to tags
df["Leisure_trip"] = df.Tags.apply(lambda tag: 1 if "Leisure trip" in tag else 0)
df["Couple"] = df.Tags.apply(lambda tag: 1 if "Couple" in tag else 0)
df["Solo_traveler"] = df.Tags.apply(lambda tag: 1 if "Solo traveler" in tag else 0)
df["Business_trip"] = df.Tags.apply(lambda tag: 1 if "Business trip" in tag else 0)
df["Group"] = df.Tags.apply(lambda tag: 1 if "Group" in tag or "Travelers with friends" in tag else 0)
df["Family_with_young_children"] = df.Tags.apply(lambda tag: 1 if "Family with young children" in tag else 0)
df["Family_with_older_children"] = df.Tags.apply(lambda tag: 1 if "Family with older children" in tag else 0)
df["With_a_pet"] = df.Tags.apply(lambda tag: 1 if "With a pet" in tag else 0)
```

```python
df.to_csv(r"../data/Hotel_Reviews_Filtered.csv", index = False)
```
